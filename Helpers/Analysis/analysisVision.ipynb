```python
import glob
import pandas as pd
import os

outputs = r'C:\Users\silvi\Documents\UFSCar\TCC\Python\My-repositories\test1\Outputs\*'
solvers = {}

for executions in glob.glob(outputs):
    rootKey = os.path.basename(executions)
    solvers[rootKey] = {}
    for subfolder in glob.glob(fr'{executions}\*\*'):
        key = os.path.basename(os.path.dirname(subfolder))
        df = pd.read_csv(fr'{subfolder}\Data\markerPose_Err.csv', index_col=0)
        df['Type'] = key
        solvers[rootKey][key] = {
            'df': df,
            'describe': df.describe()
        }
    solvers[rootKey]['concat'] = pd.concat([solver['df'] for solver in solvers[rootKey].values()])
    solvers[rootKey]['mean'] = solvers[rootKey]['concat'].groupby(['Type']).mean()
    solvers[rootKey]['min'] = solvers[rootKey]['mean'].min()
```
